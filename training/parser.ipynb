```python
from base64 import decode
import requests
from bs4 import BeautifulSoup
count = 0
number = 0
f = open('link_nums.txt', 'r', encoding="utf-8")
it = open('input.txt', 'w', encoding="utf-8")
list = f.read().split()
try:
    for i in list:
        url = f'https://www.kinoafisha.info/movies/{i}/'
        number = i
        response = requests.get(url)
        if response.status_code != 200:
            continue
        soup = BeautifulSoup(response.text, features='html.parser')
        des = soup.find("div", class_="visualEditorInsertion filmDesc_editor more_content")
        tit = soup.find("h1", class_="trailer_title")
        if des is None:
            continue
        if des.p is None:
            continue
        print(url)
        description = ""
        for t in des:
            description += t.get_text().replace('\n', ' ')
        description = description.replace('\r', '')
        description = description.replace(chr(160), '')
        description = description.replace('  ', ' ')
        it.write("<s>Название:" + tit.get_text()[:-6] + "\n")
        it.write("Описание:" + description + "<\s>\n")
        count += 1
except:
    print("Ended early on ")
    print(number)
print("dataset size =", end=" ")
print(count)
f.close()
it.close()
```
